## Inverse Z-Transform 

```python
import os
import torch
import pickle
from tqdm import tqdm
from dataset import NoisedReceptorPeptideDataset
```

```python
SCALER = "./z_score_scaler.pkl"
SPLIT_FILE = "./data/data_split.pt"
OUTPUT_FOLDER = "./data/sample_res/sample_test_{random_seed}"
```

```python
data_split = torch.load(SPLIT_FILE)
test_ds = NoisedReceptorPeptideDataset(dataset=data_split["test"])
train_val_ds = NoisedReceptorPeptideDataset(dataset=data_split["val"]+data_split["train"])
with open(SCALER, "rb") as f:
    scaler = pickle.load(f)
```

```python
for i in range(10):
    folder_path = OUTPUT_FOLDER.format(random_seed=i)
    for d in tqdm(test_ds):
        pred_emb = torch.load(os.path.join(folder_path, f"{d['pdb_id']}.pkl"))
        pred_emb = scaler.inverse_transform(pred_emb)
        torch.save(pred_emb, os.path.join(folder_path, f"{d['pdb_id']}_transformed.pt"))
```

```python
OUTPUT_FOLDER = "./data/sample_res/sample_train_val_{random_seed}"
for i in range(10):
    folder_path = OUTPUT_FOLDER.format(random_seed=i)
    for d in tqdm(train_val_ds):
        pred_emb = torch.load(os.path.join(folder_path, f"{d['pdb_id']}.pkl"))
        pred_emb = scaler.inverse_transform(pred_emb)
        torch.save(pred_emb, os.path.join(folder_path, f"{d['pdb_id']}_transformed.pt"))
```
